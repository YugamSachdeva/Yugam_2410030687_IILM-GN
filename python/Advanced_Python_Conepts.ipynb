# 1)Decorators

A decorator is a function that takes another function, wraps it inside a new function (wrapper), adds extra behavior, and returns that new function.

```python
def decorator(func):
    def wrapper():
        print('I am about to execute this func.')
        func()
        print('Func executed successfully.')
    return wrapper
@decorator  #implements the func using recorator func
def say_hello():
    print('Hello!')
say_hello()    
    
```

```python
#using decorator with arguments.
def repeat(n):#n->argument
    def decorator(func):
        def wrapper(a):
            for i in range(n):
                func(a)
        return wrapper
    return decorator

@repeat(7)
def say_hello(a):
    print(f'Hello {a}')
say_hello('dear!')
```

# 2)Getters and Setters

Getters and setters are methods used to access and modify the values of class attributes in a controlled way.

- **Getter** → used to access (get) the value  
- **Setter** → used to modify (set) the value  

In Python:
- `@property` is used to define a **getter**
- `@<attribute>.setter` is used to define a **setter** to modify the value

```python
#without getter and setter
class Employee:
    def __init__(self,name,sal):
        self.name=name
        self.sal=sal

    def first_name(self):
        l=self.name.split(" ")
        return l[0]

    def set_first_name(self,a):
        l=self.name.split(" ")
        new_name=f'{a} {l[1]}'
        self.name=new_name

e=Employee('jack doe',45896)
print(e.name)
print(e.first_name())
e.set_first_name('john')
print(e.name)
print(e.first_name())
```

```python
#with getter and setter
class Employee:
    def __init__(self,name,sal):
        self.name=name
        self.sal=sal
    @property
    def first_name(self):
        l=self.name.split(" ")
        return l[0]
        
    @first_name.setter
    def first_name(self,a):
        l=self.name.split(" ")
        new_name=f'{a} {l[1]}'
        self.name=new_name

e=Employee('jack doe',45896)
print(e.name)
print(e.first_name)
e.first_name='john'
print(e.name)
print(e.first_name)
```

# 3)Types of Methods in Python

| Method Type      | Definition |
|-----------------|------------|
| Instance Method | Operates on object data, uses self parameter to access and modify instance variables specific to each object. |
| Class Method    | Works with class-level data, uses cls parameter, and can modify shared attributes across all instances of the class. |
| Static Method   | Does not use self or cls, acts like a regular function, used for utility logic related to the class. |

# Key Differences Between Method Types

| Method Type     | Requires `self`? | Requires `cls`? | Can Access Instance Attributes? | Can Modify Class Attributes? |
|----------------|-----------------|-----------------|--------------------------------|------------------------------|
| Instance Method | Yes             | No              | Yes                            | Yes (indirectly)             |
| Class Method    | No              | Yes             | No (directly)                  | Yes                          |
| Static Method   | No              | No              | No                             | No                           |

```python
class Employee:
    company='HP'
    def __init__(self,name,sal):
        self.name=name
        self.sal=sal

    #INSTANCE METHOD:
    def info(self):
        print(f'The name is {self.name} and salary is {self.sal}.')

    #STATIC METHOD
    @staticmethod
    def sum(a,b):
        return a+b

    #CLASS METHOD
    @classmethod
    def pr_company(cls):
        print(cls.company)

    @classmethod
    def change_comp(cls,comp):
        cls.company=comp

e1=Employee('yash',4523)
e2=Employee('vikas',45223)
e1.info()#instance method
e2.info()#insatnce method
e1.pr_company () #class method
e1.change_comp('acer')#class method
e1.pr_company ()#class method
print(e1.sum(5,2))#static method
```

# 4)Magic / Dunder Methods

Magic methods, also called dunder (double underscore) methods, are special methods in Python that have double underscores at the beginning and end of their names. They define how objects behave with built-in Python features like operators, functions, and language constructs.

---

## 🔹 Purpose of Magic Methods
- Customize how objects are created and initialized  
- Enable operator overloading (like +, -, ==, etc.)  
- Provide readable and official string representations of objects  
- Control how attributes are accessed or modified  
- Allow objects to behave like functions (callable)  
- Make objects behave like containers (support length, indexing, etc.)  
- Support context management (used with `with` statement)  

---

## 🔹 Important Concepts

### 1. Object Initialization
Used to set up object data when it is created.

---

### 2. String Representation
- `__str__` → user-friendly output (used in print)  
- `__repr__` → detailed/debug representation  
- If `__str__` is missing, Python uses `__repr__`  

---

### 3. Length Behavior
Allows objects to work with functions like `len()` by defining what "length" means.

---

### 4. Operator Overloading
Lets objects work with operators like:
- Arithmetic → +, -, *, /  
- Comparison → ==, <, >, <=, >=  

---

## 🔹 One-line Summary
Magic methods allow you to **control and customize how objects behave in Python in a natural and built-in way.**

```python
class Employee:
    company='HP'
    def __init__(self,name,sal):
        self.name=name
        self.sal=sal

    def __str__(self):
        return f'name:{self.name}\nsalary:{self.sal}'

    def __len__(self):
        return len(self.name)
        
e=Employee('yash',4523)
print(str(e))
print(len(e))
```

```python
#2_--
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Person({self.name}, {self.age})"  # User-friendly

    def __repr__(self):
        return f"Person(name='{self.name}', age={self.age})"  # Unambiguous, for debugging

p = Person("Alice", 30)
print(str(p))    # Person(Alice, 30)
print(repr(p))   # Person(name='Alice', age=30)
print(p)          # Person(Alice, 30)  # print() uses __str__ if available
```

```python
#3_--
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    def __len__(self):
        return self.pages

b = Book("Python 101", 250)
print(len(b))  # 250
```

```python
#4_--
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)
    
    def __mul__(self, scalar):
      return Vector(self.x * scalar, self.y * scalar)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2  # Calls __add__
print(v3)    # Vector(6, 8)
v4 = v3 - v1
print(v4)      # Vector(4, 5)
v5 = v1 * 5
print(v5) # Vector(10, 15)
```

# 5)Exception Handling and Custom Errors

Exceptions are errors that occur during program execution and interrupt the normal flow. Python handles these using try-except blocks, allowing programs to manage errors gracefully instead of crashing.

Custom exceptions are user-defined errors created to represent specific conditions in a program, making error handling more clear and meaningful.

```python
#EXCEPTION HANDLING:-
def div(a,b):
    try:
        res=a/b

    except ZeroDivisionError:
        return ("cant div by 0")

    except ValueError:
        return ("used bad datatype for execution")

    except Exception as e:
        return ("Unknown error occured!",e)

    else:
        print("executes when no error is there.")
        return res

    finally:
        print("This block is always executed.")

print(div(4,0))
```

```python
def decorator(func):
    def wrapper(*args,**kwargs):
        print("the function is being called...")
        return func(*args, **kwargs)
    return wrapper

@decorator
def add(*args):
    return sum(args)

@decorator
def pr_info(**kwargs):
    for i in kwargs.keys():
        print(f'{i}:{kwargs[i]}')
        
print(add(0,1,2))
pr_info(name='yugam',age=20)
```
