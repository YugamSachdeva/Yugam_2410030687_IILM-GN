```python
#Read File:-
f=open("File_handle.txt","r")
content=f.read()
print(content)
f.close()
```

```python
#Write File
f=open("newfile.txt","w")
string='''I am writing a new file as i am learning how to handle file.'''
f.write(string)
f.close()
```

```python
#Append File:-
f=open("newfile.txt","a")
string='''I am appending text in a new file\n I am learning this from Python Bootcamp.'''
f.write(string)
f.close()
```

```python
#Reading file line by line:-
try:
    f=open("newfile.txt","r")
    for line in f:
        print(line)
    f.close()
except FileNotFoundError:
    print("File not found.")
```

```python
#File handling using with:-
with open("newfile.txt","r") as f:
    content=f.read()
    print(content)
    #this automatically close the file 
```

```python
#OS Modules:-
import os
a=os.listdir("dir")#Print all files name in the directory
print(a)
print(os.getcwd())#return the current directory in which working on.
os.remove("sample.txt")# deletes the file
os.rmdir("dir")#deletes the empty directories
```

```python
#Shutil Module:-
import shutil
shutil.rmtree("dir")#remove/del full directories
shutil.copy("sample.txt","john.txt")#copy the content of one file in another file
shutil.move("john.txt","dir/")#move file to directory
```

```python
#command line utilies:-
#argparse is a built-in Python module used to create command-line interfaces,It helps your program 
#accept arguments from the terminal in a clean and structured way.
import argparse

parser = argparse.ArgumentParser(description="Simple Calculator")

parser.add_argument("num1", type=float, help="First Number")
parser.add_argument("num2", type=float, help="Second Number")
parser.add_argument("Operator", choices=["add", "sub", "mul", "div"], help="Operator to perform")

args = parser.parse_args()

if args.Operator == "add":
    print(args.num1 + args.num2)

elif args.Operator == "sub":
    print(args.num1 - args.num2)

elif args.Operator == "mul":
    print(args.num1 * args.num2)

elif args.Operator == "div":
    if args.num2 == 0:
        print("Cannot divide by zero")
    else:
        print(args.num1 / args.num2)

```
