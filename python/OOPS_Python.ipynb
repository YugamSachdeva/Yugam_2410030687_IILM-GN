```python
#OOPS-Object Oriented Programming
#Class-Blueprint of Object
#Object-Instance of class

class Employee:#class
    company="HP"
    def get_sal(self):
        return 34000
e=Employee()#object
print(e.company)
print(e.get_sal())
#self-a way to refrence the object of the class being created.    
```

```python
#Constructors-Help to initialize the object data
#instance attribute=unique for each object
#class attribute=shared by all objects
class Employee:#class
    company="HP"# class attribute
    def __init__(self,name,sal,bond,company):#constructor
        self.name=name
        self.sal=sal
        self.bond=bond
        self.company=company#instance attribute
        
    def get_sal(self):
        return self.sal
        
    def get_info(self):
        print(f'The name of employee is{self.name}, his salary is {self.sal} and signed a bond for {self.bond} years.')
e1=Employee('yash',54000,4,'HP')#object
print(e1.company)
print(e1.get_sal())

e2=Employee('kartik',20000,5,'Tesla')
e2.get_info()
print(e2.company)#instance attribute
print(Employee.company)# class attribute
```

```python
#Inheritance & Polymorphism

class animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        print("Make Sound...")

class dog(animal):#inheritance
    def speak(self):#polymorphism
        super().speak()# super() method-used to call method or construvtor of parent class.
        print("WOOF!")

D=dog('vicky')
D.speak()
```

```python
#operator overloading:-
class point:
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def sum(self,p):
        return point(self.x +p.x,self.y +p.y)

    def __add__(self,p):#operator overloading --add-- helps to use(+) operator.
        return point(self.x +p.x,self.y +p.y)
        
    def prt_pnt(self):    
        print(f'x is {self.x} and y is {self.y}')
p1=point(6,3)
p2=point(3,5)
#p=p1.sum(p2)  #using sum method
#p.prt_pnt()

p=p1+p2
p.prt_pnt()
```

Create a class **Car** that has a method **drive()** which prints **"Car is moving"**.

# 📌 Constructor and Attributes

Create a class **Person** with a constructor `__init__` that accepts **name** and **age** as arguments and stores them as instance attributes.
Then create an object of the class and print the person’s **name** and **age**.


```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
    def info(self):
        print(f'The name of person is {self.name} and age is {self.age}.')
p=Person('yashu',20)
p.info()
```

# 📌 Simple Inheritance

## 🔹 Question

Create a base class **Animal** with a method `sound()` that prints **"Some sound"**.

Create a derived class **Dog** that overrides `sound()` to print **"Bark!"**.

Create an object of the Dog class and call the `sound()` method.


```python
class animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        print("Make Sound...")

class dog(animal):#inheritance
    def speak(self):#polymorphism
        super().speak()# super() method-used to call method or construvtor of parent class.
        print("WOOF!")

D=dog('vicky')
D.speak()
```
