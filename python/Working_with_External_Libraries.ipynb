### 1) Virtual Environment

A virtual environment is an isolated Python setup for a project.
It contains:

* Python interpreter
* Installed libraries

#### Why use it?

* Avoid conflicts between projects
* Use different library versions safely
* Keeps projects clean and organized

#### Creating Environment

```
python3 -m venv my_env
```

#### Activate Environment

**Windows:**

```
my_env\Scripts\activate
```

**Mac/Linux:**

```
source my_env/bin/activate
```

#### Deactivate

```
deactivate
```

### 3) Requests Module - Working with APIs

* The **requests** library is used to make HTTP requests in Python
* It helps in interacting with web APIs (Application Programming Interfaces)
* Simplifies sending and receiving data over the internet

#### Install Requests

```
pip install requests
```

#### Example (GET Request)

```
import requests

response = requests.get("https://api.example.com/data")

print(response.status_code)
print(response.text)
```

#### Common Uses

* Fetch data from APIs
* Send data to servers
* Work with JSON responses

#### Important Methods

* `requests.get()` → Get data
* `requests.post()` → Send data
* `requests.put()` → Update data
* `requests.delete()` → Delete data


````python
import requests
url = "https://api.github.com/users/octocat"  # Example API endpoint
response = requests.get(url)
if response.status_code == 200:
    data = response.json()  # Parse the JSON response
    print(data["name"])  # Access data from the JSON
else:
    print(f"Error: {response.status_code}")
# Making a POST request (for sending data to an API):
# data = {"key": "value"}
# response = requests.post(url, json=data)  # Sends data as JSON
# Other HTTP methods: put(), delete(), etc.
````

### 3) Regular Expressions in Python

* Regular expressions (regex) are used for pattern matching in strings
* Python provides regex support using the `re` module

#### Common Regex Parts

| Part  | Meaning                                                   |
| ----- | --------------------------------------------------------- |
| `\b`  | Word boundary (ensures full word match)                   |
| `\w+` | One or more word characters (letters, digits, underscore) |
| `\b`  | Word boundary (ensures complete words)                    |


````python
import re
text="The quick brown for jumps over the lazy dog."
#search for a pattern:-
match=re.search("brown",text)
if match:
    print("Match found!")
    print("start ind:",match.start())
    print("End idx:",match.end())
````

````python
import re
text = "The quick brown fox jumps over the lazy dog."
match = re.findall("the", text, re.IGNORECASE)
print("Matches:", match)
````

````python
#replace pattern:-
new_text=re.sub("dog","cat",text)
print("New text:",new_text)
````

### 5) Multithreading in Python

* Multithreading allows a program to perform multiple tasks concurrently
* It helps improve performance, especially for I/O-bound tasks

#### Multithreading (using threading module)

* Python provides multithreading support using the `threading` module


````python
import threading
import time
def worker(num):
    print(f'Thread{num}:Starting')
    time.sleep(2)
    print(f'Thread{num}:finishing')

threads=[]
for i in range(3):
    thread=threading.Thread(target=worker,args=(i,))
    threads.append(thread)
    thread.start()

for thread in threads:   
    thread.join()#wait for all threads to finish

print("All threads completed")
````
